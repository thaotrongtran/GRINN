```python
path=Path('data')
```

```python
path.ls()
```

```python
#This does not work
def get_data(bs, size):
    src = ObjectItemList.from_folder(path/'data_from_our_dataset')
    src = src.split_by_rand_pct(0)
    src = src.transform(get_transforms(), size=size)
    return src.databunch(path=path, bs=bs).normalize(imagenet_stats)
```

```python
tfms = [rotate(degrees=90, p=1.0),rotate(degrees=90, p=1.0)]
```

```python
data=ImageDataBunch.from_folder(path=path/'data_from_our_video', ds_tfms=tfms , valid_pct=0, size=256, bs=8).normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(10,10))
```

```python
learn = load_learner(path/'models')
```

```python
f=open(path/"Anno/list_category_cloth.txt", "r")
contents =f.read().splitlines()
```

```python
classes = []
for i in range(2,len(contents)):
    classes.append(contents[i].split(" ")[0])
print(classes)
```

## Offset by 2

```python
arr = learn.predict(img)[2].numpy()
k = 3
arr_top_k = np.argpartition(arr, len(arr) - k)[-k:]
```

```python
for i in arr_top_k:
    print(classes[i-1])
```

```python
for i in range(20):
    z = np.random.randint(0,len(data.train_ds))
    img  = data.train_ds[z][0]
    display(img)
    arr = learn.predict(img)[2].numpy()
    k = 5
    arr_top_k = np.argpartition(arr, len(arr) - k)[-k:]
    print(arr_top_k)
    for k in arr_top_k:
       print(classes[k])
    print("===")
```

```python
%pylab inline
from fastai.vision import *
from pathlib import Path
import pandas as pd
import json
```

```python
path=Path('data')
```

```python
path.ls()
```

```python
f=open(path/"fashion-data/train.txt", "r")
train_content =f.read().splitlines()

f=open(path/"fashion-data/test.txt", "r")
test_content =f.read().splitlines()
```

```python
images = []
for i in range(0,len(train_content)):
    file_path = train_content[i]+'.jpg'
    category = file_path.split('/')[0]
    is_valid = False;
    images.append([file_path, category, is_valid])  
    
for i in range(0,len(test_content)):
    file_path = test_content[i]+'.jpg'
    category = file_path.split('/')[0]
    is_valid = True;
    images.append([file_path, category, is_valid])
```

```python
images_df = pd.DataFrame(images)
```

```python
images_df.head()
```

```python
images_df.rename(columns = {0:'filename',1:'label', 2:'is_valid'}, inplace = True) 
```

```python
images_df.head()
```

```python
#trained_labels = labels_df.groupby('label').apply(lambda x: x.sample(n=500, replace=True))
trained_labels = images_df.sample(n=30000, random_state = 1)
```

```python
trained_labels.shape
```

```python
def get_data(bs=8, size=256):
    src = ImageList.from_df(trained_labels, path/'fashion-data/images')
    src = src.split_from_df(col=2)
    #src = src.split_by_rand_pct(0.2)
    src = src.label_from_df(cols='label')
    src = src.transform(get_transforms(flip_vert = False, max_rotate = 20, max_zoom = 1.25, max_warp = None, p_affine = 0.1, p_lighting = 0.5, ), size=size)
    return src.databunch(path=path, bs=bs).normalize(imagenet_stats)
```

```python
data = get_data()
```

```python
data
```

```python
data.show_batch(rows=3, figsize=(10,10))
```
